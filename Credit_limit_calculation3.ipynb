```python
import pandas as pd

# First, perform a left join to find rows from merged_df5 that are NOT in df_sale
merged_result = pd.merge(
    merged_df5_cleaned,             # The main dataframe (merged_df5)
    df_sale,                # The dataframe to merge with (df_sale)
    on='APON_Code',         # Merge on the 'APON_Code' column
    how='left',             # Perform a left join to keep all rows from merged_df5
    indicator=True          # Add a column to indicate where each row came from
)

# Filter rows that are only in merged_df5 and not in df_sale, and Tenure >= 60
final_result = merged_result[(merged_result['_merge'] == 'left_only') & (merged_result['Tenure'] >= 60)]

# Drop the '_merge' column as it's no longer needed
final_result = final_result.drop(columns=['_merge'])

# Display the result
final_result.head()

# Check the number of rows in the final result
print(len(final_result))
```
